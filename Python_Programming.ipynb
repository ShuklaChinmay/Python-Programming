## Doubly Linked List

**--> Introduction**

```python
class DLLNode:
  def __init__(self,data):
    self.data = data
    self.prev = None
    self.next = None

# Create a node
head = DLLNode(10)
n2  = DLLNode(20)
n3  = DLLNode(30)
n4  = DLLNode(40)

# Manually linking the linked list
head.next = n2
n2.prev = head
n2.next = n3
n3.prev = n2
n3.next = n4
n4.prev = n3

# 10 <--> 20 <--> 30 <--> 40
```

**--> Printing Of the Doubly Linked List**

```python
def print_DLL(head):
    current = head
    while current is not None:
        print(current.data, end=" <--> ")
        current = current.next
    print("None")
```

```python
# print the DOubly linked list
print_DLL(head)
```

**--> Insertion of the node at the beginning of the doubly linked list**

```python
# Logic
'''
newNode = DLLNode(5)
newNode.next = head
head.prev = newNode
head = newNode
'''

# code in the fucntion format
def insert_at_Beginning(head , value):
  newNode = DLLNode(value)
  newNode.prev = None
  newNode.next = head
  head.prev = newNode
  head = newNode
  return head
```

```python
begining_value = insert_at_Beginning(head,5)
# print the DLL
print_DLL(begining_value)
```

**--> Insertion of the node at the END of the doubly linked list**

```python
# logic
'''
# Create a new node
newNode = DLLNode(35)

# traverse till the last node
current = head
while current.next != None:
  current = current.next

# pointer adjustments
current.next = newNode
newNode.prev = current
newNode.next = None
'''

# code in the fucntion format
def insert_at_End(head , value):
  newNode = DLLNode(value)
  current = head
  while current.next != None:
    current = current.next
  current.next = newNode
  newNode.prev = current
  newNode.next = None
  return head
```

```python
head = insert_at_Beginning(head,5)
end_value = insert_at_End(head , 45)
print_DLL(end_value)
```

**--> Deletion of the node at the Beginning of the doubly linked list**

```python
# logic
'''
temp = head.next
temp.prev = None
head = temp
        OR
head.next.prev = None
head = head.next
'''

# code in the fucntion format
def delete_first(head):

  # Empty List
  if head is None:
    return None

  # Only One Node
  if head.next is None:
    return None

  #  More than one node
  temp = head.next
  temp.prev = None

  return temp
```

```python
print_DLL(delete_first(head))
```

**--> Deletion of the node at the Last of the doubly linked list**

```python
# logic
'''
# traversal to the last node
current = head
while current.next != None:
  current = current.next

# pointer adjustments
current.pre.next = None
current.pre = None
'''

# code in the function format
def delete_last(head):

  # Empty List
  if head is None:
    return None

  # Only One Node
  if head.next is None:
    return None

  current = head

   # taversal to the last node
  while current.next != None:
    current = current.next

  # pointer adjustments
  current.prev.next = None
  current.prev = None
  return head
```

```python
print_DLL(delete_last(head))
```
