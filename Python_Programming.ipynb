## Doubly Linked List

**--> Introduction**

```python
class DLLNode:
  def __init__(self,data):
    self.data = data
    self.prev = None
    self.next = None

# Create a node
head = DLLNode(10)
n2  = DLLNode(20)
n3  = DLLNode(30)
n4  = DLLNode(40)

# Manually linking the linked list
head.next = n2
n2.prev = head
n2.next = n3
n3.prev = n2
n3.next = n4
n4.prev = n3

# 10 <--> 20 <--> 30 <--> 40
```

**--> Insertion of the node at the beginning of the doubly linked list**

```python
# Logic
'''
newNode = DLLNode(5)
newNode.next = head
head.prev = newNode
head = newNode
'''

# code in the fucntion format
def insert_at_Beginning(head , value):
  newNode = DLLNode(value)
  newNode.prev = None
  newNode.next = head
  head.prev = newNode
  head = newNode
  return head
```
